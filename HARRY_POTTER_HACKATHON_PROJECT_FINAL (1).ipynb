## Let's look at the most used words in the different Harry Potter movie scripts through utilizing NLP!

![title](img/page.png)

```python
movie1 = script_one['Sentence'].values
```

```python
movie3 = script_three['SENTENCE'].values
```

```python
#Breaking our data into sentences
def get_sent_tokens(data):
    """Sentence tokenization"""
    sentences = []
    for sent in data:
        sentences.extend(sent_tokenize(sent))
    print('\nSentence tokens:', sentences)
    return sentences
```

```python
SorcerersStoneSentences = get_sent_tokens(movie1)
PrisonerOfAzkabanSentences = get_sent_tokens(movie3)
```

```python
#Breaking our sentences into words 
def get_word_tokens(data):
    '''Word tokenization'''
    words = []
    for sent in data:
        words.extend(word_tokenize(sent))
    print('\nWord tokens:', words)
    return(words)
```

```python
SorcerersStoneWords = get_word_tokens(movie1)
PrisonerOfAzkabanWords = get_word_tokens(movie3)
```

```python
#Important to clean our data before extracting key words to base visualizations off of

import re
nltk.download('stopwords')
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

def get_clean_sentences(sentences, remove_digits=False):
    '''Cleaning sentences by removing special characters and optionally digits'''
    clean_sentences = []
    for sent in sentences:
        pattern = r'[^a-zA-Z0-9\s]' if not remove_digits else r'[^a-zA-Z\s]' 
        clean_text = re.sub(pattern, '', sent) # Applying regex filter
        clean_text = clean_text.lower()  # Converting to lower case
        clean_sentences.append(clean_text)
    print('\nClean sentences:', clean_sentences)
    return clean_sentences

def filter_stopwords(words):
    '''Removing stopwords from given words'''
    filtered_words = [w for w in words if w not in stop_words]
    print('\nFiltered words:', filtered_words)
    return filtered_words
```

```python
SorcerersStoneSentences = get_clean_sentences(SorcerersStoneSentences)
SorcerersStoneWords = get_word_tokens(SorcerersStoneSentences)
SorcerersStoneWords = filter_stopwords(SorcerersStoneWords)
```

```python
PrisonerOfAzkabanSentences = get_clean_sentences(PrisonerOfAzkabanSentences)
PrisonerOfAzkabanWords = get_word_tokens(PrisonerOfAzkabanSentences)
PrisonerOfAzkabanWords = filter_stopwords(PrisonerOfAzkabanWords)
```

### Now that we've cleaned all our data, we can use our list of words to plot a frequency graph of the top 25 words spoken in Harry Potter and the Sorcerer's Stone and in Harry Potter and the Prisoner of Azkaban.

```python
import matplotlib
from nltk.probability import FreqDist
# matplotlib.use('TkAgg') 

def plot_freq_dist(words, num_words = 20):
    '''Frequency distribution'''
    fdist = FreqDist(words)
    fdist.plot(num_words, cumulative=False)
```

```python
%matplotlib inline
plot_freq_dist(SorcerersStoneWords, num_words=25)
```

```python
plot_freq_dist(PrisonerOfAzkabanWords, num_words=25)
```

### Our predictions: 

We predict that Harry Potter and the Prisoner of Azkaban will have more of a negative score than Harry Potter and the Sorcerer's Stone because the tone is more grim and less whimsical. Therefore, we expect the compound score for the first movie to be higher than that of the third movie. 

```python
def sentiment_analyzer_scores(text):
    score = analyser.polarity_scores(text) 
    print(text) 
    print('---------------------------------')
    print('MOVIE SCRIPT SCORE:')
    print('')
    print(score) 
    print('')
```
